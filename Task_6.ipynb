```python
# Smart Parking Lot Management System
def manage_parking(capacity, logs):

    current_vehicles = 0 
    peak_usage = 0  
    for log in logs:
        # if vehicle enters, increase count
        if log == "IN":
            current_vehicles += 1
        # if vehicle exits, decrease count
        elif log == "OUT":
            current_vehicles -= 1
        if current_vehicles > peak_usage:
            peak_usage = current_vehicles
    print("Currently Parked Vehicles:", current_vehicles)

    # checking parking status
    if current_vehicles > capacity:
        print("Parking Status: Full")
        print("Alert: Parking capacity exceeded!")
    else:
        print("Parking Status: Available")
    print("Peak Parking Usage:", peak_usage)

capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
manage_parking(capacity, vehicle_logs)
```

```python
# Online Food Delivery Time Estimator
def estimate_delivery_time(distance, traffic, weather):
    time = distance * 5
    # adding extra time based on traffic
    if traffic == "High":
        time += 10
    elif traffic == "Medium":
        time += 5
    # adding extra time based on weather condition
    if weather == "Rainy":
        time += 5
    elif weather == "Stormy":
        time += 15
    print("Estimated Delivery Time:", time, "minutes")

estimate_delivery_time(8, "High", "Rainy")
```

```python
# Movie Theatre Seat Occupancy Analyzer
def analyze_occupancy(total_seats, booked_seats):
    # counting how many seats are booked
    booked_count = len(booked_seats)
    # calculating occupancy percentage
    occupancy = (booked_count / total_seats) * 100
    print(f"Occupancy: {int(occupancy)}%")
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")
    # suggesting new shows if occupancy is very high
    if occupancy >= 90:
        print("Suggestion: Open additional shows")

total_seats = 200
booked_seats = [1] * 150
analyze_occupancy(total_seats, booked_seats)
```

```python
# Cloud Server Load Classification System
def classify_server_load(cpu_readings):
    # calculating average CPU load
    average = sum(cpu_readings) / len(cpu_readings)
    # printing average load
    print(f"Average CPU Load: {int(average)}%")
    # checking server status based on average load
    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

cpu = [45, 60, 70, 85, 90]
classify_server_load(cpu)
```

```python
# Smart Classroom Resource Usage Monitor
def monitor_resources(resource_usage):
    overused = []  
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)
    # printing overused resources and energy alert
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
monitor_resources(usage)
```

```python
# Online Event Registration Capacity Controller
def manage_registrations(capacity, registrations):
    confirmed = 0  
    waitlisted = 0  
    for i in range(registrations):
        # if capacity is not full, confirm registration
        if confirmed < capacity:
            confirmed += 1
        # if capacity is full, add to waitlist
        else:
            waitlisted += 1
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    if waitlisted > 0:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")

manage_registrations(100, 105)
```
