```python
# Problem 1: Smart Light Controller
# Problem Statement
# Create a class to control a smart light that can be turned ON or OFF and display its status.
# üåç Real-Time Use
# Smart home automation systems
# üì• Sample Input
# Light Name: Bedroom Light
# Action: ON
# üì§ Expected Output
# Bedroom Light is ON
```

```python
class SmartLight:
    
    # Constructor
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"   # Default status

    # Method to turn ON the light
    def turn_on(self):
        self.status = "ON"

    # Method to turn OFF the light
    def turn_off(self):
        self.status = "OFF"

    # Method to display status
    def display_status(self):
        print(f"{self.light_name} is {self.status}")


# Main Program 
light_name = input("Light Name: ")
action = input("Action (ON/OFF): ")

light = SmartLight(light_name)

if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()

light.display_status()
```

```python
# Problem 2: Employee ID Card System
#  Problem Statement
# Create a class that stores employee name, ID, and department and displays ID card details.
# Real-Time Use
# Corporate HR systems
# üì• Sample Input
# Employee Name: Rahul
# Employee ID: EMP102
# Department: IT
# üì§ Expected Output
# Employee ID Card
# Name: Rahul
# ID: EMP102
# Department: IT
```

```python
class Employee:

    # Constructor
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    # Method to display ID card details
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Main Program
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")

emp = Employee(name, emp_id, department)
emp.display_id_card()
```

```python
# Problem 3: Mobile Contact Record
# Problem Statement
# Create a class to store contact name and phone number and display contact information.
# Real-Time Use
# Mobile contact applications
# üì• Sample Input
# Contact Name: Anita
# Phone Number: 9876543210
# üì§ Expected Output
# Contact Saved
# Name: Anita
# Phone: 9876543210
```

```python
class Contact:

    # Constructor
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    # Method to display contact details
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# ---- Main Program ----
name = input("Contact Name: ")
phone = input("Phone Number: ")

contact = Contact(name, phone)
contact.display_contact()
```

```python
# Problem 4: Product Price Tag Generator
# Problem Statement
# Create a class that stores product name and price and prints a formatted price tag.
#  Real-Time Use
# Retail billing systems
# üì• Sample Input
# Product Name: Headphones
# Price: 2499
# üì§ Expected Output
# Product: Headphones
# Price: ‚Çπ2499
```

```python
class Product:

    # Constructor
    def __init__(self, name, price):
        self.name = name
        self.price = price

    # Method to print price tag
    def print_price_tag(self):
        print("Product:", self.name)
        print("Price: ‚Çπ" + str(self.price))


# Main Program
name = input("Product Name: ")
price = input("Price: ")

product = Product(name, price)
product.print_price_tag()
```

```python
# Problem 5: Movie Rating Display System
# Problem Statement
# Create a class that stores movie name and rating and displays movie details.
# Real-Time Use
# Streaming platforms
# üì• Sample Input
# Movie Name: Inception
# Rating: 4.8
# üì§ Expected Output
# Movie: Inception
# Rating: 4.8 / 5
```

```python
class Movie:

    # Constructor
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    # Method to display movie details
    def display_movie(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


# Main Program 
name = input("Movie Name: ")
rating = input("Rating: ")

movie = Movie(name, rating)
movie.display_movie()
```

```python
# Problem 6: Delivery Address Manager
# Problem Statement
# Create a class that stores customer name and delivery address and prints delivery details.
# Real-Time Use
# Courier and logistics platforms
# üì• Sample Input
# Customer Name: Suman
# Address: Hyderabad
# üì§ Expected Output
# Delivery Details
# Customer: Suman
# Address: Hyderabad
```

```python
class Delivery:

    # Constructor
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    # Method to print delivery details
    def display_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Main Program
name = input("Customer Name: ")
address = input("Address: ")

delivery = Delivery(name, address)
delivery.display_details()
```
